**Bài 1 – Khai báo biến cơ bản**
Khai báo hai biến a và b lần lượt nhận giá trị 10 và 3.ính và in ra:
Tổng (+) 
Hiệu (- ) 
Tích (*)
Thương (/) 


```python
a = 10
b = 3

tong = a + b
hieu = a - b
tich = a * b
thuong = a / b

print("a =", a, "| b =", b)
print("Tổng:", tong)
print("Hiệu:", hieu)
print("Tích:", tich)
print("Thương:", f"{thuong:.2f}")

```

**2. Cho hai biến x = 17, y = 5. In ra:**
Kết quả chia nguyên x // y 
Phần dư x % y 



```python

x = 17
y = 5

chia_nguyen = x // y
chia_du = x % y

print("x =", x, "| y =", y)
print("Chia nguyên x//y:", chia_nguyen)
print("Chia dư x%y:", chia_du)
```

**Bài 3 – Toán tử lũy thừa**
Khai báo biến n = 4.
Tính:

```python
n = 4
binh_phuong = n ** 2
lap_phuong = n ** 3

print("n =", n)
print("n^2 =", binh_phuong)
print("n^3 =", lap_phuong)
```

**Bài 4 – So sánh hai số** 
Cho a = 15, b = 20. In ra kết quả của các biểu thức:
a == b
a != b # so sánh khác nhau 
a > b
a < b



```python
a = 15
b = 20

print("a =", a, "| b =", b)
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b :", a > b)
print("a < b :", a < b)
```

**Bài 5 – Kiểm tra số chẵn**
Cho biến num. Viết biểu thức kiểm tra num có chia hết cho 2 hay không (chỉ dùng toán tử, không dùng if).



```python
num = 42  # bạn có thể đổi giá trị để thử

is_even = (num % 2 == 0)

print("num =", num)
print("num là số chẵn?", is_even)
```

**Bài 6 – Toán tử logic**
Cho:
a = 10
b = 5
c = 20
Tính và in ra kết quả của:
(a > b) and (c > a)
(a < b) or (c > a)
not (a == b)


```python
a = 10
b = 5
c = 20

expr1 = (a > b) and (c > a)
expr2 = (a < b) or (c > a)
expr3 = not (a == b)

print("a =", a, "| b =", b, "| c =", c)
print("(a > b) and (c > a):", expr1)
print("(a < b) or (c > a) :", expr2)
print("not (a == b)        :", expr3)
```

**Bài 7 – Gán và cập nhật giá trị** 
Khai báo x = 10. Lần lượt thực hiện:
Tăng x lên 5
Nhân x với 2
Giảm x đi 3 In giá trị x sau mỗi bước.



```python
x = 10
print("Ban đầu x =", x)
x += 5 # tăng x thêm 5
print("Sau khi tăng 5: x =", x)
x *= 2 #nhân x với 2
print("Sau khi nhân 2: x =", x)
x -= 3 # giảm x đi 3
print("Sau khi giảm 3: x =", x)
```

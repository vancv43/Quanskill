```python
# Daily sales data for a week
daily_sales = [1200, 800, 1500, 900, 1800, 2200, 1600]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Basic statistics
total_sales = sum(daily_sales)
average_sales = total_sales / len(daily_sales)
max_sales = max(daily_sales)
min_sales = min(daily_sales)

print(f"Total sales: ${total_sales}")        # Output: Total sales: $10000
print(f"Average sales: ${average_sales:.2f}")    # Output: Average sales: $1428.57
print(f"Max sales: ${max_sales}")            # Output: Max sales: $2200
print(f"Min sales: ${min_sales}")            # Output: Min sales: $800

# Find best selling day
best_day_index = daily_sales.index(max_sales)
print(f"Best day: {days[best_day_index]}")   # Output: Best day: Sat
print ("\n")
# Danh sch nhn vin
employees = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"]
print(employees)
```

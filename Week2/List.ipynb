1. In điểm vị trí thứ 1, vị trí thứ 4 và vị trí cuối cùng.
a. in điểm từ : index 1 -> 3 [1:4] 
b. in điểm 3 sinh viên đầu tiên [:3] 
c. in điểm từ vị trí 4 đến hết danh sách [4:] 
d. in điểm vị trí chẵn [::2]

```python
student_scores = [78, 85, 92, 60, 88, 95, 70]
print("Student Scores:", student_scores)# In danh sách điểm số của sinh viên. 
print("Highest Score:", max(student_scores)) # In điểm số cao nhất. 
print("Lowest Score:", min(student_scores))  # In điểm số thấp nhất.
print("Average Score:", sum(student_scores) / len(student_scores))  # In điểm số trung bình.
house_price =[250,150,300,250,200]
# danh sách có 5 phần tử, tính từ vị trí [0] đến vị trí [n-1]
print("vị trí đầu tiên=",house_price[0])
print("vị trí cuối cùng=",house_price[-1])
#Yêu cầu in ví số 3 của danh sanh sách ? 
print("vị trí thứ 3=",house_price[2])
#Yêu cầu in ví số 3 của danh sanh sách  --- IGNORE ---

#slicing 
print("giá nhà từ vị trí 1 đến vị trí 3:",house_price[1:4])
print("giá nhà từ đầu đến vị trí 2:",house_price[:3])
print("giá nhà từ vị trí 2 đến cuối:",house_price[2:])
print("giá nhà toàn bộ danh sách:",house_price[:])
#slicing  --- IGNORE ---

```

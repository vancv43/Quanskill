Y√äU C·∫¶U:

2.L·ªçc d·ªØ li·ªáu nv thu·ªôc ph√≤ng IT.? 
3 Th·ªëng k√™ l∆∞∆°ng (T√≠nh t·ªïng l∆∞∆°ng, l∆∞∆°ng trung b√¨nh) c·ªßa c√¥ng ty.?
4.C·∫≠p nh·∫≠t l∆∞∆°ng nh∆∞ng KH√îNG s·ª≠a tuple
    - T·∫°o list m·ªõi updated_employees
    - Nh√¢n vi√™n ph√≤ng IT tƒÉng l∆∞∆°ng 10%
    - C√°c ph√≤ng kh√°c gi·ªØ nguy√™n

    - Th√™m nh√¢n vi√™n m·ªõi, in l·∫°i danh s√°ch nv sau khi th√™m. 



```python
employees = [
    ("E001", "Nguyen Van A", "IT", 1500),
    ("E002", "Tran Thi B", "HR", 1200),
    ("E003", "Le Van C", "IT", 1800),
    ("E004", "Pham Thi D", "Finance", 2000),
    ("E005", "Hoang Van E", "IT", 1700)
]

#1.Duy·ªát danh s√°ch nh√¢n vi√™n
print("\n=== 1. DUY·ªÜT DS NH√ÇN VI√äN ===")
for emp_id, name, dept, salary in employees:
    print(f"ID: {emp_id}, Name: {name}, Department: {dept}, Salary: {salary}")

#2.L·ªçc nh√¢n vi√™n IT c√≥ l∆∞∆°ng > 1600
print("\n=== 2. NH√ÇN VI√äN PH√íNG IT ===")
for emp_id, name, dept, salary in employees:
    if dept == "IT":
        print(f"{emp_id} | {name} | {salary}")

#3.TH·ªêNG K√ä L∆Ø∆†NG NV CTY 
print("\n=== 3. T·ªîNG L∆Ø∆†NG V√Ä L∆Ø∆†NG TRUNG B√åNH ===")
total_salary = sum(salary for emp_id, name, dept, salary in employees)
avg_salary = total_salary / len(employees) # trong python kh√¥ng c√≥ h√†m avg n√™n ta t·ª± t√≠nh 
print(f"T·ªïng l∆∞∆°ng: {total_salary}")
print(f"L∆∞∆°ng trung b√¨nh: {avg_salary}")

#4. C·∫≠p nh·∫≠t l∆∞∆°ng nh∆∞ng KH√îNG s·ª≠a tuple,T·∫°o list m·ªõi updated_employees
#IT tƒÉng 10%
#Ph√≤ng kh√°c gi·ªØ nguy√™n
#employees ban ƒë·∫ßu kh√¥ng ƒë·ªïi

print("\n=== 4. C·∫¨P NH·∫¨T L∆Ø∆†NG NH√ÇN VI√äN ===")
updated_employees = []
for emp_id, name, dept, salary in employees:
    if dept == "IT":
        new_salary = salary * 1.10
    else:
        new_salary = salary

    # t·∫°o tuple m·ªõi
    updated_employees.append(
        (emp_id, name, dept, new_salary)
    )

print("\n=== DANH S√ÅCH SAU KHI C·∫¨P NH·∫¨T L∆Ø∆†NG ===")
for emp in updated_employees:
    print(emp)
#5. TTh√™m nh√¢n vi√™n m·ªõi & in l·∫°i danh s√°ch
print("\n=== 5. TH√äM NH√ÇN VI√äN M·ªöI ===")
new_employee = ("E006", "Nguy·ªÖn Th·ªã Di·ªÖm H∆∞∆°ng", "Marketing", 1600)
updated_employees.append(new_employee)
print("\n=== DANH S√ÅCH SAU KHI TH√äM NH√ÇN VI√äN M·ªöI ===")
for emp_id, name, dept, salary in updated_employees:
    print(f"{emp_id} | {name} | {dept} | {salary}")

```

**D·∫†NG B√ÄI T·∫¨P: ƒê·ªäNH NGHƒ®A H√ÄM TRONG DOANH NGHI·ªÜP**
1. Th·ªëng k√™ l∆∞∆°ng nh√¢n vi√™n, l∆∞∆°ng th·∫•p nh·∫•t, cao nh·∫•t v√† trung b√¨nh c·ªßa c√¥ng ty.
2. Vi·∫øt h√†m salary_stats(salaries),tr·∫£ v·ªÅ tuple (min_salary, max_salary, avg_salary). 
M√¥ t·∫£: def salary_stats(salaries):#del : ƒë·ªãnh nghƒ©a (define)

3.Th·ªëng k√™ doanh s·ªë b√°n h√†ng c·ªßa t·ª´ng nh√¢n vi√™n trong m·ªôt k·ª≥ (tu·∫ßn / th√°ng).
Theo m√£:  M√£ nh√¢n vi√™n, Danh s√°ch doanh s·ªë c√°c ƒë∆°n h√†ng. 
    ƒê·ªãnh nghƒ©a h√†m calculate_sales_stats(sales)
    H√†m nh·∫≠n v√†o list doanh s·ªë
Tr·∫£ v·ªÅ tuple g·ªìm:
    T·ªïng doanh s·ªë
    Doanh s·ªë cao nh·∫•t
    Doanh s·ªë trung b√¨nh
    In b√°o c√°o doanh s·ªë cho nh√¢n vi√™n

```python
salaries = [1500, 1200, 1800, 2000, 1700]

def salary_stats(salary_list):
    min_salary = min(salary_list)
    max_salary = max(salary_list)
    avg_salary = sum(salary_list) / len(salary_list)
    return min_salary, max_salary, avg_salary

min_s, max_s, avg_s = salary_stats(salaries)

print(f"Min: {min_s}, Max: {max_s}, Average: {avg_s}")
print("S·ªë nh√¢n vi√™n:", len(salaries))
```

```python
#3.Th·ªëng k√™ doanh s·ªë b√°n h√†ng c·ªßa t·ª´ng nh√¢n vi√™n trong m·ªôt k·ª≥ (tu·∫ßn / th√°ng).
# ƒê·ªãnh nghƒ©a h√†m t√≠nh doanh s·ªë
def calculate_sales_stats(sales):
    total_sales = sum(sales)
    max_sales = max(sales)
    avg_sales = total_sales / len(sales)
    return total_sales, max_sales, avg_sales

# D·ªØ li·ªáu doanh s·ªë c·ªßa m·ªôt nh√¢n vi√™n
employee_id = "E007"
monthly_sales = [1200, 1800, 1500, 2000, 1700]

# G·ªçi h√†m ƒë√£ ƒë·ªãnh nghƒ©a
total, max_val, avg_val = calculate_sales_stats(monthly_sales)
# In b√°o c√°o
print(f"Employee ID: {employee_id}")
print(f"Total Sales: {total}")
print(f"Max Sale: {max_val}")
print(f"Average Sales: {avg_val}")
print("Number of orders:", len(monthly_sales))
```
